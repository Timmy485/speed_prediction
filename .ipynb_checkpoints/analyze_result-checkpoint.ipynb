```python
fileout2 = "/srv/beegfs02/scratch/aegis_guardian/data/speed_prediction/logs/deep_train548668.out"
with open(fileout2, "r") as file:
    textout2 = file.readlines()
```

```python

import re
train_l2 = [float(
        v.strip("\n").split()[-4])
            for v in textout2 if re.findall("val_loss:", v)]
val_l2 = [float(
        v.strip("\n").split()[-1])
            for v in textout2 if re.findall("val_loss:", v)]
epochs2 = list(range(1, len(val_l2)+1))
print(train_l2)
print(val_l2)
print(epochs2)
```

```python
import matplotlib.pyplot as plt
plt.plot(epochs2[1:], train_l2[1:])
plt.plot(epochs2[1:], val_l2[1:])
# plt.ylim([0.000001, 0.06])
plt.legend(["train", "val"])
plt.show()
```
